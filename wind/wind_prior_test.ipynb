```python
import numpy as np
import matplotlib.pyplot as plt

# Domain
Lx = 1.0
Ly = 1.0

# Grid
Nx = 25
Ny = 25

x = np.linspace(0, Lx, Nx)
y = np.linspace(0, Ly, Ny)
X, Y = np.meshgrid(x, y)

# Number of modes
r = 2

# Random coefficients
sigma = 1.0 # assumed constant across modes, can be varied
a = np.random.normal(0, sigma, size=(r, r))
b = np.random.normal(0, sigma, size=(r, r))

# Initialize velocity fields
u = np.zeros_like(X)
v = np.zeros_like(X)

# Basis expansion
for i in range(1, r+1):
    for j in range(1, r+1):

        phi = np.cos(i*np.pi*Y/Ly) * np.cos(j*np.pi*X/Lx)
        psi = np.sin(i*np.pi*Y/Ly) * np.cos(j*np.pi*X/Lx)

        u += a[i-1, j-1] * phi
        v += b[i-1, j-1] * psi

# Plot
plt.figure(figsize=(8,8))
plt.quiver(X, Y, u, v)
plt.title("Sampled Wind Velocity Field")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
